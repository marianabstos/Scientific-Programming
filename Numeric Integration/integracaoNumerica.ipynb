```python
import numpy as np
import math
```

# Funções derivadas das equações A, B e C

```python
derivada_eq_a = lambda x: np.exp(x)
derivada_eq_b_2nd = lambda x: -(x/np.sqrt(1-x**2))
derivada_eq_b_4th = lambda x: -(1/(1-x**2)**(3/2)) 
derivada_eq_c_2nd = lambda x: -(2*np.exp(-x**2)*x)
derivada_eq_c_4th = lambda x: np.exp(-x**2)*(4*(x**2)-2)
```

# Fórmulas de Newton-Cotes

```python
def simpson(f,a,b):
    aux_1 = (b-a)
    aux_2 = (a+b)/2
    
    resp = aux_1 * ((f(a)+4*f(aux_2)+f(b))/6)
    
    return resp
```

```python
def retangulo(f,a,b):
    aux_1 = (b-a)
    aux_2 = (a+b)/2
    
    resp = aux_1 * f(aux_2)
    
    return resp
```

```python
def trapezoidal(f,a,b):
    aux_1 = (b-a)
    
    resp = (f(a) + f(b))/2
    
    return resp
```

# Cálculo erro de aproximação

```python
def erro_simpson(derivada_4th,a,b):
    values = []
    try:
        values.append(derivada_4th(a))
    except:
        pass
    
    try:
        values.append(derivada_4th(b))
    except:
        pass
    
    derivada = max(values)
    
    return abs(-(((b-a)**5)/2880)*derivada)
```

```python
def erro_retangulo(derivada_2nd,a,b):
    values = []
    
    try:
        values.append(derivada_2nd(a))
    except:
        pass
    
    try:
        values.append(derivada_2nd(b))
    except:
        pass
    
    derivada = max(values)
    
    return abs(((float((b-a))**3)/24)*derivada)
```

```python
def erro_trapezio(derivada_2nd,a,b):
    values = []
    
    try:
        values.append(derivada_2nd(a))
    except:
        pass
    
    try:
        values.append(derivada_2nd(b))
    except:
        pass
    
    derivada = max(values)
    
    return abs(-((float((b-a))**3)/12)*derivada)
```

# Método de Quadratura Adaptativa

```python
def calculo_quadratura(f,a,b,eps,result,tipo,eq):
    m = (a+b)/2
    
    if tipo == 'simpson':
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print(tipo)
        
        if eq == 'a':
            erro_esquerda = erro_simpson(derivada_eq_a, a,m)
            erro_direita = erro_simpson(derivada_eq_a, m,b)
        elif eq == 'b':
            erro_esquerda = erro_simpson(derivada_eq_b_4th, a,m)
            erro_direita = erro_simpson(derivada_eq_b_4th, m,b)
        elif eq == 'c':
            erro_esquerda = erro_simpson(derivada_eq_c_4th, a,m)
            erro_direita = erro_simpson(derivada_eq_c_4th, m,b)
        
        esquerda = simpson(f,a,m)
        print("a = {0}, m = {1}, resultado = {2}, erro = {3}".format(a,m,esquerda,erro_esquerda))
        direita = simpson(f,m,b)
        print("m = {0}, b = {1}, resultado = {2}, erro = {3}".format(m,b,direita, erro_direita))
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        
    if tipo == 'retangulo':
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print(tipo)
        
        if eq == 'a':
            erro_esquerda = erro_retangulo(derivada_eq_a, a,m)
            erro_direita = erro_retangulo(derivada_eq_a, m,b)
        elif eq == 'b':
            erro_esquerda = erro_retangulo(derivada_eq_b_2nd, a,m)
            erro_direita = erro_retangulo(derivada_eq_b_2nd, m,b)
        elif eq == 'c':
            erro_esquerda = erro_retangulo(derivada_eq_c_2nd, a,m)
            erro_direita = erro_retangulo(derivada_eq_c_2nd, m,b)
        
        esquerda = retangulo(f,a,m)
        print("a = {0}, m = {1}, resultado = {2}, erro = {3}".format(a,m,esquerda,erro_esquerda))
        direita = retangulo(f,m,b)
        print("m = {0}, b = {1}, resultado = {2}, erro = {3}".format(m,b,direita, erro_direita))
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    
    if tipo == 'trapezio':
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print(tipo)
        
        if eq == 'a':
            erro_esquerda = erro_trapezio(derivada_eq_a, a,m)
            erro_direita = erro_trapezio(derivada_eq_a, m,b)
        elif eq == 'b':
            erro_esquerda = erro_trapezio(derivada_eq_b_2nd, a,m)
            erro_direita = erro_trapezio(derivada_eq_b_2nd, m,b)
        elif eq == 'c':
            erro_esquerda = erro_trapezio(derivada_eq_c_2nd, a,m)
            erro_direita = erro_trapezio(derivada_eq_c_2nd, m,b)
        
        esquerda = trapezoidal(f,a,m)
        print("a = {0}, m = {1}, resultado = {2}, erro = {3}".format(a,m,esquerda,erro_esquerda))
        direita = trapezoidal(f,m,b)
        print("m = {0}, b = {1}, resultado = {2}, erro = {3}".format(m,b,direita, erro_direita))
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    
    if (result - esquerda - direita) < eps:
        return esquerda + direita
    
    return calculo_quadratura(f,a,m,eps,esquerda,tipo,eq) + calculo_quadratura(f,m,b,eps,direita,tipo,eq) 
```

```python
def quadratura_simpson(f,a,b,eps,eq):
    return calculo_quadratura(f,a,b,eps,simpson(f,a,b),"simpson",eq)
```

```python
def quadratura_retangulo(f,a,b,eps,eq):
    return calculo_quadratura(f,a,b,eps,retangulo(f,a,b),"retangulo",eq)
```

```python
def quadratura_trapezio(f,a,b,eps,eq):
    return calculo_quadratura(f,a,b,eps,trapezoidal(f,a,b),"trapezio",eq)
```
